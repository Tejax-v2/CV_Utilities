# On Grayscale

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

img_path = Path("images") / "cat.jpg"

# Reading the image as grayscale
image = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)

# Histogram equalization
equalized_img = cv2.equalizeHist(image)

# Plotting the figures
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Equalized Image")
plt.imshow(equalized_img, cmap="gray")
plt.axis("off")

plt.show()
```

# RGB Images with RGB Channels

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

img_path = Path("images") / "nature.jpg"

# Reading the image as RGB (since opencv reads channels as BGR by default)
image = cv2.cvtColor(cv2.imread(str(img_path)), cv2.COLOR_BGR2RGB)

# Extracting individual channels
red = cv2.extractChannel(image, coi=0)
green = cv2.extractChannel(image, coi=1)
blue = cv2.extractChannel(image, coi=2)

# Equalizing individual channels
equalized_red = cv2.equalizeHist(image[:,:,0])
equalized_green = cv2.equalizeHist(image[:,:,1])
equalized_blue = cv2.equalizeHist(image[:,:,2])

# Merging equalized channels
equalized_img = cv2.cvtColor(cv2.merge([equalized_blue, equalized_green, equalized_red]), cv2.COLOR_RGB2BGR)

# Plotting the figures
plt.figure(figsize=(10,16))
plt.subplot(4,2,1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(4,2,2)
plt.title("Equalized Image")
plt.imshow(equalized_img, cmap="gray")
plt.axis("off")

plt.subplot(4,2,3)
plt.title("Original Red")
plt.imshow(red, cmap="gray")
plt.axis("off")

plt.subplot(4,2,4)
plt.title("Equalized Red")
plt.imshow(equalized_red, cmap="gray")
plt.axis("off")

plt.subplot(4,2,5)
plt.title("Original Green")
plt.imshow(green, cmap="gray")
plt.axis("off")

plt.subplot(4,2,6)
plt.title("Equalized Green")
plt.imshow(equalized_green, cmap="gray")
plt.axis("off")

plt.subplot(4,2,7)
plt.title("Original Blue")
plt.imshow(blue, cmap="gray")
plt.axis("off")

plt.subplot(4,2,8)
plt.title("Equalized Blue")
plt.imshow(equalized_blue, cmap="gray")
plt.axis("off")

plt.show()
```

# RGB Images with YCrCb Channels

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

img_path = Path("images") / "nature.jpg"

# Reading image as RGB
image = cv2.cvtColor(cv2.imread(str(img_path)), cv2.COLOR_BGR2RGB)

# Converting to YUV color space (Y:Luminance, UV:Chrominance)
image_yuv = cv2.cvtColor(image, cv2.COLOR_RGB2YUV)

# Extracting luminance channel
lum = image_yuv[:,:,0]

# Histogram equalization
equalized_lum = cv2.equalizeHist(lum)

# Merging the channels
equalized_img_yuv = cv2.merge([equalized_lum,image_yuv[:,:,1],image_yuv[:,:,2]])

# Converting back to RGB
equalized_img = cv2.cvtColor(equalized_img_yuv, cv2.COLOR_YUV2RGB)

# Plotting the figures
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Equalized Image")
plt.imshow(equalized_img, cmap="gray")
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Original Luminance")
plt.imshow(lum, cmap="gray")
plt.axis("off")

plt.subplot(2,2,4)
plt.title("Equalized Luminance")
plt.imshow(equalized_lum, cmap="gray")
plt.axis("off")

plt.show()
```

**NOTE:** OpenCV reads image channels in the order BGR while plt reads in the order RGB. To adapt to this sequence change we need to change the channels order for RGB images.
